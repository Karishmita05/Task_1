```python
import pandas as pd
```

```python
data=pd.read_csv("../data/raw/netflix_titles(dataset3).csv",index_col=False)
```

```python
#check for missing values
data.isnull().sum()
```

```python
#handle missing value
data['director'] = data['director'].fillna('Unknown')
data['cast'] = data['cast'].fillna('Unknown')
data['country'] = data['country'].fillna('Unknown')
data['rating'] = data['rating'].fillna('Not Rated')
```

```python
#handle date column
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
```

```python
#check for duplicates
data.duplicated().sum()
```

```python
#standardized text columns
data['type'] = data['type'].str.upper()
data['country'] = data['country'].str.title()
```

```python
#handle duration column
data[['duration_value', 'duration_unit']] = data['duration'].str.split(' ', expand=True)
data['duration_value'] = pd.to_numeric(data['duration_value'], errors='coerce')
data['duration_unit'] = data['duration_unit'].str.lower()
```

```python
#feature engineering
data['year_added'] = data['year_added'].astype('Int64')
data['month_added'] = data['month_added'].astype('Int64')
```

```python
data.to_csv("../data/processed/dataset4_clean.csv",index=False)
```
