```python
import pandas as pd
```

```python
data=pd.read_csv("../data/raw/Medical_Appointment_No_Shows(dataset2).csv",index_col=False)
```

```python
data.head()
```

```python
data.info()
```

```python
#check missing values
data.isnull().sum() 
```

```python
#check duplicate rows
data.duplicated().sum() 
```

```python
#fix column names
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
```

```python
#convert data types
data['scheduledday'] = pd.to_datetime(data['scheduledday'])
data['appointmentday'] = pd.to_datetime(data['appointmentday'])
data['age'] = data['age'].astype(int)
data['patientid'] = data['patientid'].astype('int64').astype(str)
```

```python
#standardize text values
data['gender'] = data['gender'].str.upper()
data['no_show'] = data['no_show'].str.lower()
data['no_show'] = data['no_show'].map({'yes': 1, 'no': 0})
```

```python
#check and fix inconsistent data
data['age'].unique()
data['gender'].unique()
data['no_show'].unique()
```

```python
#Feature Engineering: adding a new column waiting_days in the dataset
data['waiting_days'] = ( data['appointmentday'].dt.floor('D') -data['scheduledday'].dt.floor('D')
).dt.days
```

```python
data.to_csv('../data/processed/dataset2_clean.csv',index=False)
```
