```python
import pandas as pd
```

```python
data=pd.read_csv("../data/raw/marketing_campaign(dataset1).csv",sep="\t")
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe() #income has missing values or null vales
```

```python
data.isnull().sum() #income has 24 null vales
```

```python
#to fix missing values of income
data['Income'].fillna(data['Income'].median(), inplace=True)
```

```python
data.isnull().sum() #missing vales handled
```

```python
data.duplicated().sum()
data.drop_duplicates(inplace=True)
```

```python
data['Dt_Customer']=pd.to_datetime(data['Dt_Customer'], dayfirst= True) #convert date
```

```python
data['Income']=data['Income'].astype(int) #convert numeric
```

```python
#standardize text data
data['Education']=data['Education'].str.lower().str.strip()
data['Marital_Status']=data['Marital_Status'].str.lower().str.strip()
```

```python
#fix inconsistent category
data['Marital_Status'].unique()
data['Marital_Status']=data['Marital_Status'].replace({'alone':'single','together':'married','absurd':'single','yolo':'single'})
```

```python
# remove useless columns
data.nunique()
data.drop(['Z_CostContact','Z_Revenue'],axis=1,inplace=True)
```

```python
#Feature Engineering
data['age']=2026 -data['Year_Birth']
data['total_children']=data['Kidhome']+data['Teenhome']
data['total_spent'] = ( data['MntWines'] + data['MntFruits'] + data['MntMeatProducts'] + data['MntFishProducts'] + data['MntSweetProducts'] + data['MntGoldProds']
)
```

```python
#rename columns
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

```python
#outlier handling
data['income'].describe()
data = data[data['income'] < 600000]
```
